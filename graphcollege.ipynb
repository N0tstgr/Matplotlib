```python
import numpy as np
import matplotlib.pyplot as plt

frequency = np.array([50,100,150,200,250,300,350,400,450,500,550,600,
                      650,700,750,800,1000,50000,100000,150000,200000,
                      210000,220000,230000,240000,250000,290000,310000,
                      330000,350000,370000,390000,410000,430000,450000,
                      470000,490000,510000,530000,550000,570000,590000])

Vout = np.array([0.32,0.56,0.88,1.44,1.92,2.40,2.96,3.28,3.52,4.08,
                 4.24,4.40,4.40,4.48,4.48,4.48,4.48,4.48,4.48,4.48,
                 4.40,4.32,4.32,4.24,4.24,4.24,4.16,4.08,4.16,4.08,
                 3.92,3.84,3.76,3.68,3.60,3.52,3.36,3.28,3.20,
                 3.12,3.04,2.96])

vin = 2
gain = Vout / vin

plt.figure()
plt.semilogx(frequency, gain, marker='o')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (Vout/Vin)")
plt.title("Multistage Amplifier")
plt.grid(True)
plt.show()
```
