```python
import numpy as np
import matplotlib.pyplot as plt

# Frequency in Hz
frequency = np.array([
    50,100,150,200,250,300,350,400,450,500,
    550,600,700,800,900,1000,1100,1200,1300,
    1400,1500,1600,1700,1800,1900,2000,
    3000,4000,5000,6000,7000,8000
])

# Output Voltage in Volts
vout = np.array([
    0.32,0.48,0.64,0.72,0.96,1.36,1.36,1.28,1.28,1.36,
    1.44,1.44,1.68,1.76,1.84,1.92,2.00,2.08,2.16,
    2.16,2.24,2.24,2.24,2.24,2.32,2.32,
    2.40,2.48,2.56,2.56,2.56,2.56
])

plt.figure()
plt.semilogx(frequency, vout, marker='o')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Output Voltage (V)")
plt.title("Single Stage Amplifier Frequency Response")
plt.grid(True)

plt.show()
```
