```python
#plt.plot(x,y, color = 'color name', linestyle='line_Style', linewidth=value, marker='marker symbol', label='label name')
import matplotlib.pyplot as plt
month = [2,4,5,6,7]
sales = [500,3000,30000,202222,333344]
plt.plot(month, sales, color='blue', linestyle='--', linewidth = 2, marker = 'o', label='2026 Sales data')
plt.xlabel("Months")
plt.ylabel("sales per month")
plt.title("Monthly Sales Data Report")
plt.legend(loc='upper left', fontsize=13)
plt.grid(color='green', linestyle=':', linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.show()
```
